```python

#1.Accept project marks, internal marks, external marks from student
#2.passmarks is 50 in each subject
# 3.Calculate total score if student got more than or equal to 50 in each subject .

# total_score=70% on project +20% on external =10% on internal
# Print A grade if score is 90+
#       B grade if score b/w 70-90
#       C grade if score b/w 50-70*/
project=int(input("Enter project marks:"))        #90
internal=int(input("Enter internal marks:"))      #78
external=int(input("Enter external marks:"))      #45
if project>=50 and internal>=50 and external>=50:
  total=(project*70)/100+(external*20)/100+(external*10)/100
  print("Total score is:",total)
  if total>=90:
    print("A grade")
  elif total>=70:
    print("B grade")
  else:
    print("C grade")
  print("hi")
else:
  print("hello")




```
